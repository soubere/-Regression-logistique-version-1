```python
#Phase 1e importation des donnees
import pandas as pd
titanic = pd.read_csv('titanic-passengers.csv', sep = ';', encoding = "ISO-8859-1")
titanic.head()
df = titanic.copy()
df.head()
```

```python
#Pretraitement rechercher des valeurs manquantes et remplacer les donner manquante
df.isnull().sum()
```

```python
#remplacer les donner manquante de lage
frequentvalue = df['Age'].value_counts().idxmax()
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
#remplacer les donner manquante de Cabin
frequentCabin = df['Cabin'].value_counts().idxmax()
df["Cabin"].fillna('G6',inplace=True)
df.isnull().sum()
```

```python
#remplacer les donner manquante de Embarked
frequentEmbarked = df['Embarked'].value_counts()
df["Embarked"].fillna('S',inplace=True)
df.isnull().sum()
```

```python
df['Survived'].values
```

```python
dictSurvived = {'Yes':1, 'No':0}
df['Survived'].replace(dictSurvived, inplace=True)
df.head(8)
```

```python
df.isnull().sum()
```

```python
dictSex = {'male':1, 'female':0}
df['Sex'].replace(dictSex, inplace=True)
df.head(8)
```

```python
#import relevant libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

#Appliquons la regression logistique

x = df[['Sex', 'Age']]
y = df['Survived']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0) 

regreslogist = LogisticRegression()   #build our logistic model
regreslogist.fit(x_train, y_train)  #fitting training data
y_prediction  = regreslogist.predict(x_test)    #testing modelâ€™s performance
```

```python
print("Accuracy={:.2f}".format(regreslogist.score(x_test, y_test)))
```
